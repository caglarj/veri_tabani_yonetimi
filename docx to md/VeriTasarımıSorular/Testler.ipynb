```markdown

11. İlişkisel veritabanı modelinde bir veritabanı tablosunun birincil anahtarı (primary key) ne işe yarar?

A) Tablodaki tüm verileri siler.
B) Tablodaki tüm verileri günceller.
C) Tablodaki tüm verileri sıralar.
D) Tablodaki her bir kaydın benzersiz bir şekilde tanımlanmasını sağlar.
E) Tablodaki verileri yedekler.

----

cevap : D) Tablodaki her bir kaydın benzersiz bir şekilde tanımlanmasını sağlar.

Birincil anahtar, bir ilişkisel veritabanı tablosundaki her bir kaydın benzersiz bir şekilde tanımlanmasını sağlayan bir sütundur. Bu sütun, her bir kaydın diğer kayıtlardan ayırt edici bir değere sahip olmasını sağlar. Birincil anahtar, tablodaki verileri silmek, güncellemek, sıralamak veya yedeklemek gibi işlemler için kullanılmaz. Ancak, birincil anahtarın bir kaydı benzersiz şekilde tanımlaması nedeniyle, tablodaki verilerin doğru bir şekilde yönetilmesine ve ilişkilendirilmesine yardımcı olur.
```
